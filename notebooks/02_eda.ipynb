```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv(r"C:\Users\arysh\projectHotel\cleaned_data.csv")
df.head()

```

```python
df.describe()
```

```python
df.groupby("distribution_channel").agg({
    "adr": "sum",
    "lead_time": "mean"
})
```

```python
df.groupby("hotel")["is_canceled"].mean()  *100
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
sns.boxplot(x="is_canceled",y="lead_time", data=df)
plt.title ("lead time of the cancilation")
plt.show()
```

```python
month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
df['arrival_date_month'] = pd.Categorical(df['arrival_date_month'], categories=month_order, ordered=True)
monthly_bookings = df['arrival_date_month'].value_counts().sort_index()

monthly_bookings.plot(kind='bar', title='Bookings per Month')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='barh', title='Top 10 Countries by Booking Count')
plt.xlabel('Bookings')
plt.ylabel('Country')
plt.show()
```

```python
cancel_by_segment = df.groupby('market_segment')['is_canceled'].mean().sort_values()
cancel_by_segment.plot(kind='barh', title='Cancellation Rate by Market Segment')
plt.xlabel('Cancellation Rate')
plt.show()
```

```python
sns.boxplot(x='is_canceled', y='adr', data=df)
plt.title('ADR vs Cancellation')
plt.show()
```

```python
sns.histplot(df['stays_in_week_nights'], bins=30, kde=True)
plt.title('Stay Length Distribution')
plt.xlabel('Nights')
plt.ylabel('Frequency')
plt.show()
```

```python
df.groupby('stays_in_weekend_nights')['is_canceled'].mean().plot(kind='bar', title='Cancellation Rate: Weekend Stay')
plt.xticks(ticks=[0, 1], labels=['No Weekend Stay', 'Weekend Stay'], rotation=0)
plt.ylabel('Cancellation Rate')
plt.show()
```

```python
df.groupby('is_business')['is_canceled'].mean().plot(kind='bar', title='Cancellation Rate: Business vs Leisure')
plt.xticks(ticks=[0, 1], labels=['Leisure', 'Business'], rotation=0)
plt.ylabel('Cancellation Rate')
plt.show()
```

```python
sns.barplot(x='total_special_requests', y='is_canceled', data=df)
plt.title('Cancellation Rate by Number of Special Requests')
plt.ylabel('Cancellation Rate')
plt.xlabel('Number of Special Requests')
plt.show()
```
